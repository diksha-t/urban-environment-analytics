```python
import pandas as pd 
import os
import sqlalchemy as text
from sqlalchemy import create_engine

df = pd.read_csv("D:\\urban-environment-analytics\\data\\BandraMumbaiMPCB.csv")
df.head()
df.info()
df.isnull().sum()

```

```python
columns_to_drop = [
    "Unnamed: 0",
    "To Date",
    "NO",
    "NOx",
    "Ozone",
    "Benzene",
    "Toluene",
    "WS",
    "WD",
    "SR",
    "BP",
    "VWS",
    "RF",
    "TOT-RF"
]
df = df.drop(columns=columns_to_drop, errors="ignore")
df.head()
df.info()
```

```python
df = df.rename(columns={"From Date": "timestamp"})
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df.info()
```

```python
ordered_columns = [
    "timestamp",
    "PM2.5",
    "PM10",
    "NO2",
    "SO2",
    "CO",
    "Temp",
    "RH"  
]

df = df[ordered_columns]
```

```python
df.to_csv("D:\\urban-environment-analytics\\data\\BandraMumbaiMPCB.csv", index=False)
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace(".", "", regex=False)
      .str.replace("-", "_")
)

df.columns.tolist()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sensor_cols = ['pm25', 'pm10', 'no2', 'so2', 'co', 'temp', 'rh']

df = df.dropna(subset=sensor_cols, how='all')

for col in sensor_cols:
    df[col] = df[col].fillna(df[col].median()) # fills remaining missing values with median
    df = df[df[col] >= 0] # removes negative values
    
df.isnull().sum()
```

```python
df.shape
```

```python
os.makedirs("data/processed", exist_ok=True)

df.to_csv("data/processed/cleaned_air_quality.csv", index=False) # cleaned dataset is saved for phase 2
```

```python
df = pd.read_csv(
    "data/processed/cleaned_air_quality.csv",
    parse_dates=["timestamp"]
) # reload cleaned csv

engine = create_engine(
    "postgresql+psycopg2://postgres:201204@localhost:5432/urban_env_analytics"
) # connected to postgres

df.to_sql(
    "air_quality",
    engine,
    if_exists="append",
    index=False
) # load data into the table

print("Data loaded successfully!")
```

```python
with engine.connect() as conn:
    result = conn.execute(text("SELECT COUNT(*) FROM air_quality"))
    print(result.fetchone())
```
