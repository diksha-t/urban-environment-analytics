```python
import pandas as pd 
df = pd.read_csv("D:\\urban-environment-analytics\\data\\BandraMumbaiMPCB.csv")

df.head()
df.info()
df.isnull().sum()

```

```python
columns_to_drop = [
    "Unnamed: 0",
    "To Date",
    "NO",
    "NOx",
    "Ozone",
    "Benzene",
    "Toluene",
    "WS",
    "WD",
    "SR",
    "BP",
    "VWS",
    "RF",
    "TOT-RF"
]
df = df.drop(columns=columns_to_drop, errors="ignore")
df.head()
df.info()
```

```python
df = df.rename(columns={"From Date": "timestamp"})
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df.info()
```

```python
ordered_columns = [
    "timestamp",
    "PM2.5",
    "PM10",
    "NO2",
    "SO2",
    "CO",
    "Temp",
    "RH"  
]

df = df[ordered_columns]
```

```python
df.to_csv("D:\\urban-environment-analytics\\data\\BandraMumbaiMPCB.csv", index=False)
```
